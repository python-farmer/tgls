```python
import pandas as pd
```

```python
#In der nächsten Zeile wird in den Anführungszeichen der Dateiname der einzuspielenden Excel-Datei vermerkt
df = pd.read_excel('batchchange_excel_sechster_versuch_j.xlsx')

df
```

```python
def pica_schreiben(row):
    '''Dies ist ein Docstring'''
    pica = f'''{row.IDN_DNB}
0500\tOa
0600\tro
1131\t!94827171X!
1700\t/1XA-DE
3210\t{row.Normnummer}
3260\tTechnische Normen, Gütevorschriften und Lieferbedingungen
4000\t{row.Normnummer}, {row.Titel}
4004\t
4030\tFrankfurt am Main ; Leipzig : Deutsche Nationalbibliothek
4060\t1 Online-Ressource
4190\tDDR-Standard
'''
    pica +="\t\n"
    
    return pica
```

```python
for index, row in df.iterrows():
    print(pica_schreiben(row))
```

```python
#In der nächsten Zeile wird in den Anführungszeichen hinter dem ersten f der gewünschte Dateiname der erzeugten .dat-Datei vermerkt
with open(f"batchchange_eingabe.dat", 'w', encoding='utf-8') as f:
    for index, row in df.iterrows():
        f.write(pica_schreiben(row))
```

```python
def pica_idn(row):
    liste = str(row.IDN_DNB)
    
    liste +="\n"
    
    return liste
```

```python
#In der nächsten Zeile wird in den Anführungszeichen hinter dem ersten f der gewünschte Dateiname der erzeugten IDN-Liste vermerkt
with open(f"batchchange_idn_liste.txt", 'w', encoding='utf-8') as f:
    for index, row in df.iterrows():
        f.write(pica_idn(row))
```
